# Part 3: The awesome `NetworkX` library

In case it wasn't clear by now, this class is about YOU analyzing networks. And it wouldn't be right to start the first lecture without playing a little bit with network analysis (there will be much more on this in the following lectures). So here goes...
`NetworkX` should already be installed as part of your _Anaconda_ Python distribution. But you don't know how to use it yet. The best way to get familiar is to work through a tutorial.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

### Create Graphs

```python
# Create an empty undirected graph, a directed graph, and a multigraph
G = nx.Graph()
D = nx.DiGraph()
MG = nx.MultiGraph()

print('Graphs created:')
print(f'Undirected Graph: {G}')
print(f'Directed Graph: {D}')
print(f'MultiGraph: {MG}')
```

### Adding Nodes and Edges

```python
# Adding nodes and edges to the undirected graph
G.add_node(1)                   # Add a single node
G.add_nodes_from([2, 3, 4])     # Add multiple nodes
G.add_edge(1, 2)                # Add an edge between two nodes
G.add_edges_from([(2, 3), (3, 4)])  # Add multiple edges

# Visualize the undirected graph
plt.figure(figsize=(6, 4))
nx.draw(G, with_labels=True, node_color='lightblue', node_size=700)
plt.title("Undirected Graph")
plt.show()
```

```python
# Adding edges to the directed graph
D.add_edge(1, 2)  # Adds a directed edge from node 1 to node 2
D.add_edges_from([(2, 3), (3, 1), (3, 4)])

# Visualize the directed graph
plt.figure(figsize=(6, 4))
nx.draw(D, with_labels=True, node_color='lightgreen', node_size=700, arrows=True)
plt.title("Directed Graph")
plt.show()
```

### Neighbors and Degrees

```python
# Nodes and edges
print('Nodes in G:', G.nodes())               # Get all nodes
print('Edges in G:', G.edges())               # Get all edges

# Checking for neighbors
print('Neighbors of node 2:', list(G.neighbors(2)))  # List of neighbors of node 2

# Degree of nodes
print('Degree of node 2:', G.degree(2))            # Degree of node 2
print('All node degrees:', dict(G.degree()))        # Degrees of all nodes
```

### Shortest Path

```python
# Calculate and print the shortest path from node 1 to node 4
path = nx.shortest_path(G, source=1, target=4)
print('Shortest path from node 1 to node 4:', path)

# Highlight the shortest path in the graph visualization
plt.figure(figsize=(6, 4))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=700)
nx.draw_networkx_edges(G, pos, edgelist=[(path[i], path[i+1]) for i in range(len(path)-1)], width=4, edge_color='r')
plt.title("Shortest Path from Node 1 to Node 4")
plt.show()
```

### Clustering Coefficient

```python
# Calculate and display clustering coefficient of node 2
clustering = nx.clustering(G, 2)
overall_clustering = nx.average_clustering(G)

print(f'Clustering coefficient of node 2: {clustering}')
print(f'Average clustering coefficient: {overall_clustering}')
```

### Centrality Measures

```python
# Degree, Closeness, and Betweenness centrality
degree_centrality = nx.degree_centrality(G)
closeness_centrality = nx.closeness_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)

print('Degree centrality:', degree_centrality)
print('Closeness centrality:', closeness_centrality)
print('Betweenness centrality:', betweenness_centrality)
```

### Save and Load Graphs

```python
# Write graph to GML format and read it back
nx.write_gml(G, 'graph.gml')
G_loaded = nx.read_gml('graph.gml')

# Visualize the loaded graph
plt.figure(figsize=(6, 4))
nx.draw(G_loaded, with_labels=True, node_color='orange', node_size=700)
plt.title("Loaded Graph from GML")
plt.show()
```

### Generate Graphs

```python
# Generate and visualize a complete graph with 5 nodes
complete_graph = nx.complete_graph(5)
plt.figure(figsize=(6, 4))
nx.draw(complete_graph, with_labels=True, node_color='lightpink', node_size=700)
plt.title("Complete Graph with 5 Nodes")
plt.show()

# Generate and visualize a random graph using the Erdos-Renyi model
random_graph = nx.erdos_renyi_graph(10, 0.5)
plt.figure(figsize=(6, 4))
nx.draw(random_graph, with_labels=True, node_color='lightblue', node_size=700)
plt.title("Random Graph (Erdos-Renyi Model)")
plt.show()
```

### Connected Components

```python
# Identify and print connected components
components = list(nx.connected_components(G))
print('Connected components:', components)

# Extract and visualize a subgraph containing nodes 1, 2, and 3
subgraph = G.subgraph([1, 2, 3])
plt.figure(figsize=(6, 4))
nx.draw(subgraph, with_labels=True, node_color='lightgreen', node_size=700)
plt.title("Subgraph of Nodes 1, 2, and 3")
plt.show()
```
