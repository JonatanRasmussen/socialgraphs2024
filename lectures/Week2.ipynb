```python
import networkx as nx
import numpy as np
```

```python
# Ex 2.3
# Create the undirected graph (a)
G_a = nx.Graph()
G_a.add_edges_from([(1,2), (1,3), (1,4), (1,6), (2,3), (3,6)])

# Create the directed graph (b)
G_b = nx.DiGraph()
G_b.add_edges_from([(1,2), (1,4), (2,3), (3,1), (3,2), (6,1), (6,3), (6,4)])
```

```python
# Adjacency matrices
adj_matrix_a = nx.to_numpy_array(G_a)
adj_matrix_b = nx.to_numpy_array(G_b)

print("Adjacency matrix for graph (a):")
print(adj_matrix_a)
print("\nAdjacency matrix for graph (b):")
print(adj_matrix_b)
```

```python
# Link lists
link_list_a = list(G_a.edges())
link_list_b = list(G_b.edges())

print("\nLink list for graph (a):")
print(link_list_a)
print("\nLink list for graph (b):")
print(link_list_b)
```

```python
# Average clustering coefficient for graph (a)
avg_clustering_a = nx.average_clustering(G_a)
print(f"\nAverage clustering coefficient for graph (a): {avg_clustering_a}")
```

```python

# Paths of length 3 from node 1 to node 3
def paths_of_length(G, start, end, n):
    # Get the adjacency matrix
    adj_matrix = nx.adjacency_matrix(G).todense()

    # Compute A^n where A is the adjacency matrix
    paths = np.linalg.matrix_power(adj_matrix, n)

    # Return the number of paths of length n from start to end
    return int(paths[start-1, end-1])

n = 3
paths_a = paths_of_length(G_a, 1, 3, n)
paths_b = paths_of_length(G_b, 1, 3, n)

print(f"\nPaths of length 3 from node 1 to node 3 in graph (a): {paths_a}")
print(f"Paths of length 3 from node 1 to node 3 in graph (b): {paths_b}")
```

```python
# Ex 2.5
# Create the bipartite graph
B = nx.Graph()
B.add_edges_from([(1,7), (2,8), (2,9), (3,8), (3,9), (4,9), (4,10), (5,10), (5,11), (6,11)])
```

```python
# Adjacency matrix
adj_matrix_bipartite = nx.to_numpy_array(B)
print("\nAdjacency matrix for bipartite graph:")
print(adj_matrix_bipartite)
```

```python
# Projections
purple_nodes = [1, 2, 3, 4, 5, 6]
green_nodes = [7, 8, 9, 10, 11]

purple_projection = nx.bipartite.projected_graph(B, purple_nodes)
green_projection = nx.bipartite.projected_graph(B, green_nodes)

print("\nAdjacency matrix for purple nodes projection:")
print(nx.to_numpy_array(purple_projection))
print("\nAdjacency matrix for green nodes projection:")
print(nx.to_numpy_array(green_projection))
```

```python
# Average degrees
avg_degree_purple = sum(dict(B.degree(purple_nodes)).values()) / len(purple_nodes)
avg_degree_green = sum(dict(B.degree(green_nodes)).values()) / len(green_nodes)

print(f"\nAverage degree of purple nodes in bipartite graph: {avg_degree_purple}")
print(f"Average degree of green nodes in bipartite graph: {avg_degree_green}")
```

```python
# Average degrees in projections
avg_degree_purple_proj = sum(dict(purple_projection.degree()).values()) / len(purple_nodes)
avg_degree_green_proj = sum(dict(green_projection.degree()).values()) / len(green_nodes)

print(f"\nAverage degree in purple nodes projection: {avg_degree_purple_proj}")
print(f"Average degree in green nodes projection: {avg_degree_green_proj}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random
import networkx as nx
```

### Binning real numbers:

```python
# Generate 5000 data points from a Gaussian distribution
data = np.random.normal(loc=2, scale=0.125, size=5000)

# Bin the data into 10 bins
counts, bins = np.histogram(data, bins=10)

# Plot the binned data
plt.plot(bins[:-1], counts, 'o-')
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Binned Gaussian Distribution')
plt.show()
```

The np.histogram function returns two arrays: counts (the number of values in each bin) and bins (the bin edges). The bins array has one more element than counts because it includes both edges of each bin.

To plot this, we use bins[:-1] as the x-values because we want the left edge of each bin. This matches the length of counts.

### Binning integers:

```python
# Generate 5000 numbers from a Poisson distribution
poisson_data = np.random.poisson(lam=10, size=5000)

# Find min and max
min_val = np.min(poisson_data)
max_val = np.max(poisson_data)

# Create bin vector
v = np.arange(min_val, max_val + 2)

# Bin the data
counts, bins = np.histogram(poisson_data, bins=v)

# Plot as bar chart
plt.bar(bins[:-1], counts, width=1, align='edge')
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Poisson Distribution')
plt.show()

print(f"Sum of bin counts: {np.sum(counts)}")
```

The first and last elements of the binning vector v are min_val and max_val + 1 respectively. This ensures that all values, including the maximum, are included in the bins.

### Binning and plotting degree distributions

```python
# Calculate p for ER graph
N = 5000
k = 10
p = k / (N - 1)

# Generate ER graph
G = nx.erdos_renyi_graph(N, p)

# Get degree sequence
degree_sequence = [d for n, d in G.degree()]

# Bin the degree sequence
counts, bins = np.histogram(degree_sequence, bins=np.arange(min(degree_sequence), max(degree_sequence) + 2))

# Plot
plt.bar(bins[:-1], counts, width=1, align='edge')
plt.xlabel('Degree')
plt.ylabel('Count')
plt.title('Degree Distribution of ER Graph')
plt.show()
```

This should produce a plot that looks similar to the Poisson distribution from the previous part, as the degree distribution of an ER graph follows a Poisson distribution.

```python
import networkx as nx
import math
import numpy as np
import matplotlib.pyplot as plt

N = 3000
p = 1e-3

# 1. Expected number of links
L = (N * (N - 1) / 2) * p
print(f"1. Expected number of links: {L}")
```

```python
# 2. Network regime
if p < math.log(N) / N:
    print("2. The network is in the subcritical regime")
elif p > math.log(N) / N:
    print("2. The network is in the supercritical regime")
else:
    print("2. The network is at the critical point")
```

```python
# 3. Probability pc for critical point
pc = math.log(N) / N
print(f"3. Probability pc for critical point: {pc}")
```

```python
# 4. Number of nodes Ncr for one component
Ncr = 1 / p
print(f"4. Number of nodes Ncr for one component: {Ncr}")
```

```python
# 5. Average degree and average distance for network in (d)
"""
The issue lies in the calculation of the average distance. The formula used (d = math.log(Ncr) / math.log(k_cr)) is correct, but we need to be careful about the value of k_cr.
Let's break it down:
Ncr = 1 / p = 1 / 0.001 = 1000
k_cr = (Ncr - 1) * p = 999 * 0.001 = 0.999
The problem occurs because log(k_cr) is negative when k_cr is between 0 and 1, which is the case here (0.999 < 1).
To correct this, we should use the formula for the average path length in an ErdÅ‘s-RÃ©nyi network when it's in the connected regime:
d â‰ˆ log(N) / log(<k>)
Where N is the number of nodes and <k> is the average degree.
Let's correct the calculation:
"""
N = 3000
p = 1e-3
k = (N - 1) * p
d = math.log(N) / math.log(k) if k > 1 else float('inf')
print(f"5. Average degree <k>: {k}")
print(f"   Average distance <d>: {d}")
```

```python
# 6. Degree distribution (Poisson approximation)
def poisson(k, lambda_):
    return (lambda_**k * math.exp(-lambda_)) / math.factorial(k)

k_values = range(20)
pk_values = [poisson(k, L/N) for k in k_values]

plt.figure(figsize=(10, 6))
plt.bar(k_values, pk_values)
plt.xlabel('Degree (k)')
plt.ylabel('Probability (p_k)')
plt.title('Degree Distribution (Poisson Approximation)')
plt.show()
```

```python
import networkx as nx
import random
import matplotlib.pyplot as plt

# Create a random network with 200 nodes and average degree of 1.5
N = 200
p = 1.5 / (N - 1)
G = nx.erdos_renyi_graph(N, p)

# Plot the initial network
plt.figure(figsize=(12, 8))
nx.draw(G, node_size=20, node_color='#A0CBE2', with_labels=False)
plt.title("Random Network")
plt.show()
```

```python
# Extract the Giant Connected Component (GCC)
gcc = max(nx.connected_components(G), key=len)
G_gcc = G.subgraph(gcc).copy()

# Choose a random node from the GCC
start_node = random.choice(list(G_gcc.nodes()))

# Find nodes 2 steps away from the start node
path_lengths = nx.single_source_shortest_path_length(G_gcc, start_node)
nodes_2_steps = [node for node, length in path_lengths.items() if length == 2]

# Prepare node colors
node_colors = ['#A0CBE2' if node not in [start_node] + nodes_2_steps else '#ff0000' for node in G_gcc.nodes()]
node_colors[list(G_gcc.nodes()).index(start_node)] = '#000000'

# Plot the GCC with colored nodes
plt.figure(figsize=(12, 8))
nx.draw(G_gcc, node_size=50, node_color=node_colors, with_labels=False)
plt.title("Giant Connected Component with Colored Nodes")
plt.show()
```

This code creates a random network, extracts the Giant Connected Component, identifies nodes 2 steps away from a randomly chosen start node, and plots the result with appropriate coloring.
